```python
# Apply Seaborn to data

# Read data from CSV file
from pathlib import Path

filepath = 'c:\\Users\\lyleb\\Box Sync\\.PMB Master\\Engineering\\Engineering Job Folder\\'
folder = 'EPRI NMAC\\Bearing Action Advisor 2017 W985\\BAA 3_0 Elliptical bearings\\'
directory = filepath + folder
open_dir = Path(directory)
#if open_dir.is_dir():
#    print ('Named directory exists')
filename = 'Scratch effect from MAXBRG D20 0118.csv'
getfile=filepath+folder+filename
#opened_file = Path(getfile)
#if opened_file.is_file(): 
#    print ('Named file exists')
#else:
#    print (filename+': does not exist')

import csv
with open(getfile) as ff:
    reader = csv.reader(ff, delimiter=',')
    impactdata = [(float(col1),float(col2),float(col3),float(col4),float(col5),
             float(col6),float(col7),float(col8),float(col9),float(col10),float(col11))
             for col1,col2,col3,col4,col5,col6,col7,col8,col9,col10,col11 in reader]

import pandas as pd
import numpy as np
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns

cols=['Speed','ID','Length','Preload','SpecLoad','Somm','Hmin','Tmax','L/D','Ratio','Impact']
frame = pd.DataFrame(impactdata,columns=cols)
#print (frame['Somm'])
plt.plot(frame['L/D'],frame['Impact'],'g*')
plt.ylabel('Impact %')
plt.xlabel('L/D (dim)')
sns.pairplot(frame[['Somm','L/D','Impact']])
plt.show()
```
